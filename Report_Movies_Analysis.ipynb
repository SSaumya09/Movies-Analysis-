# Visualization

```python
Movies_plot = Movies_IMDb.groupby('genre')['budget'].mean()
```

```python
# Creating a line plot
plt.figure(figsize=(14,6 ))
plt.plot(Movies_plot.index, Movies_plot.values, marker='o', linestyle='-')
plt.title('BUDGET')
plt.xlabel('Year')
plt.ylabel('BOX-OFFICE')
plt.grid(True)
plt.show()
```

This Line chart shows the Box-Office on the Y-axis and Budget on the X-axis. we can infer from this graph that High budgeted movies of the genre Animation has gotten huge success in the Box-office. As we can see that Animation, Family Movies and Adventure are the most popular genres. The least box office collection are of the genres Thriller and western.


```python
# Count the number of movies in each genre
genre_counts = Movies_IMDb['genre'].value_counts()

# Select the top 8 most frequent genres
top_genres = genre_counts.nlargest(8)

# Sum the counts of the remaining genres and classify them as "Others"
other_genres_count = genre_counts[~genre_counts.index.isin(top_genres.index)].sum()
top_genres['Others'] = other_genres_count

# Define custom colors for the pie chart
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c2f0c2', '#6666ff', '#ffb366']

# Create a pie chart with custom styling
plt.figure(figsize=(6, 6))
plt.pie(top_genres, labels=top_genres.index, autopct='%1.1f%%', startangle=90, colors=colors, pctdistance=0.85)
plt.title('Genre Distribution', fontsize=14)

# Add a circle in the center to make it look like a donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')
plt.tight_layout()
plt.show()
```

This is a donut pie chart which shows the different genres of movies release in the past 30 years. The highest is action and comdey which people prefer to watch and has been trending over the years. The lowest are Horror, adventure, animation and biograohy and crime.

```python
# Filter the dataset for movies with gross and ratings
filtered_data = Movies_IMDb.dropna(subset=['gross', 'rating'])

# Group the data by rating and calculate the mean gross for each rating
rating_vs_gross = filtered_data.groupby('rating')['gross'].mean()

# Create a bar graph
plt.figure(figsize=(8, 6))
plt.bar(rating_vs_gross.index, rating_vs_gross.values, color='hotpink', edgecolor='k')
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Average Gross (in millions)', fontsize=12)
plt.title('Average Gross (in millions) vs. Rating', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Customize the tick labels
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Add a background color
ax = plt.gca()
ax.set_facecolor('#f5f5f5')

plt.show()
```

This Bar plot visually represents the relationship between Movie ratings on IMBd and their average gross earnings which helps to see how the movie ratings impact their financial success at the Box- office.

```python
# Filter out rows with missing year or genre values
df = Movies_IMDb.dropna(subset=['year', 'genre'])

# Extract the year and genre columns
year_genre_data = df[['year', 'genre']]

# Pivot the data to create a table for the stacked bar chart
pivot_table = year_genre_data.pivot_table(index='year', columns='genre', aggfunc=len, fill_value=0)

# Create a stacked bar chart
plt.figure(figsize=(12, 6))
ax = pivot_table.plot(kind='bar', stacked=True, colormap='viridis')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.title('Genre Distribution Over Time', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Customize the tick labels
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Move the legend outside of the graph
ax.legend(title='Genre', fontsize=10, title_fontsize=12, loc='upper left', bbox_to_anchor=(1, 1))

# Add a background color
ax.set_facecolor('#f5f5f5')

plt.show()
```

This is a stacked bar chart which visually represents the distribution of movie genres which has evolved over a period of time, providing insights into the trends and the changes in the film industry.

```python
# Filter out rows with missing year, budget, or gross values
df = Movies_IMDb.dropna(subset=['year', 'budget', 'gross'])

# Group the data by year and calculate the total budget and total gross for each year
budget_vs_gross = df.groupby('year')[['budget', 'gross']].sum()

# Create a dual-axis line plot
fig, ax1 = plt.subplots(figsize=(15, 6))

# Plot the budget on the first y-axis
ax1.plot(budget_vs_gross.index, budget_vs_gross['budget'], marker='o', linestyle='-', color='b', label='Budget')
ax1.set_xlabel('Year')
ax1.set_ylabel('Budget (in millions)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Create a second y-axis for gross
ax2 = ax1.twinx()
ax2.plot(budget_vs_gross.index, budget_vs_gross['gross'], marker='o', linestyle='-', color='g', label='Gross')
ax2.set_ylabel('Gross (in millions)', color='g')
ax2.tick_params(axis='y', labelcolor='g')

# Add legends for both lines
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title('Budget and Gross Over the Years', fontsize=14)
plt.grid(True)
plt.show()
```

This is a Line graph which represents Budget of the movie on the Y-axis and Budget and gross collection of the movie over the years. there is a drop in 2020 for budget and budget and gross because there were too many null values in the dataset.

```python
# Filter out rows with missing vote and genre values
df = Movies_IMDb.dropna(subset=['votes', 'genre'])

# Group the data by genre and calculate the average number of votes for each genre
genre_vs_votes = df.groupby('genre')['votes'].mean()

# Create a grouped bar chart
plt.figure(figsize=(12, 6))
genre_vs_votes.plot(kind='bar', color='purple', edgecolor='k')
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Average Votes', fontsize=12)
plt.title('Average Votes by Genre', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Customize the tick labels
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

# Add a background color
ax = plt.gca()
ax.set_facecolor('#f5f5f5')

plt.show()
```

In our analysis, we found that the “Mystery” genre garnered the highest average number of votes, indicating high audience engagement and interest in this genre. This was followed by the “Family”, “Animation” and “Action” genres, which also received significantly high average votes, highlighting their popularity among viewers. These findings suggest that films in the "Mystery", "Family", "Animation" and "Action" genres tend to resonate well with audiences, possibly due to their engaging storytelling, interesting storytelling and family content. As we consider choosing a genre for our next project, this information provides valuable guidance in choosing a genre that can maximize audience engagement and overall revenue success at the box office.
